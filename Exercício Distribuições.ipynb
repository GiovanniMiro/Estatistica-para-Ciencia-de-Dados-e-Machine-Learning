```python
def dist_normal(dataset, curve = True, grupos = 10):
    return sns.histplot(dataset, kde = curve, bins = grupos)
```

```python
def dist_normal_pad(dataset, curve = True):
    media = dataset.mean()
    dp = np.std(dataset)
    return sns.histplot((dataset - media)/dp, kde = curve, bins = 10)
```

```python
def dist_frequencia(dataset):
    return sns.countplot(dataset)
```
