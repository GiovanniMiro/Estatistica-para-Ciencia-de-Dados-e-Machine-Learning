```python
import pandas as pd
import numpy as np
from scipy import stats
import statistics
```

```python
dataset = pd.read_csv('credit_data.csv')
```

```python
dataset.isnull().sum()
```

É possível ver que existem 3 idades que não foram preenchidas

```python
nulos = dataset[dataset.isnull().any(axis=1)]
nulos
```

```python
dataset['age'].mean(), dataset['age'].median() #Colocar o () é fundamental pq indica que é para fazer a média e mediana usando todos os valores deste atributo
```

```python
dataset['age'] = dataset['age'].replace(to_replace = np.nan, value = dataset['age'].mean()) #np.nan indica que identifica onde tem valor nulo.
```

```python
dataset[dataset.isnull().any(axis=1)]
```
